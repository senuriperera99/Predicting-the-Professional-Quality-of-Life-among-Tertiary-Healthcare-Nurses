## Cts-Nominal

```python
from scipy.stats import f_oneway

# Specify response variables and the nominal variable
response_vars = ['CS_Score', 'BO_Score', 'STS_Score']
nominal_var = 'Hospital'

# Perform ANOVA for each response variable
anova_results = {}
for response_var in response_vars:
    # Extract data for each group (level of the nominal variable)
    groups = [group[1][response_var].dropna() for group in df.groupby(nominal_var)]

    # Perform ANOVA
    f_statistic, p_value = f_oneway(*groups)

    # Store results
    anova_results[response_var] = {'F-Statistic': f_statistic, 'p-value': p_value}

# Display ANOVA results
for response_var, results in anova_results.items():
    print(f"ANOVA Results for '{response_var}' by '{nominal_var}':")
    print(f"F-Statistic: {results['F-Statistic']}")
    print(f"P-Value: {results['p-value']}")

    # Interpretation
    alpha = 0.05  # significance level
    if results['p-value'] < alpha:
        print("The p-value is less than 0.05, so we reject the null hypothesis.")
        print(f"There is significant evidence that '{response_var}' varies across different levels of '{nominal_var}'.")
    else:
        print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.")
        print(f"There is not enough evidence to suggest that '{response_var}' varies across different levels of '{nominal_var}'.")
    
    print("\n")
```
